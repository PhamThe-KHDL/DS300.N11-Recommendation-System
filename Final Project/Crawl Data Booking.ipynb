### Crawl data infomation of hotel and user

**Thông tin dataset:**

| URL Hotel | Location | HotelID | Name Hotel | Descriptions | Address | UserID | User | Rating |
| ----- | ----- | ----- | ----- | ----- | ----- | ----- | ----- | ----- |
| URL của khách sạn | Địa điểm của khách sạn (Tên tỉnh/TP) | ID của khách sạn | Tên của khách sạn | Mô tả của khách sạn | Địa chỉ của khách sạn | ID của user | Tên của user (người đánh giá) | rating của user |

````python
def Crawl_Hotels_Data(url):
  content_hotel = requests.get(url).content
  soup_hotel = BeautifulSoup(content_hotel, 'html.parser')
  # URL Hotel
  url_hotel = url
  # Location
  for i in range(len(dest_id)):
    if dest_id[i] in url:
      location = dest_name[i]
  # Name Hotel
  name_hotel = soup_hotel.find('div',{'id':'hp_hotel_name'}).find('h2').text.strip()
  # Descriptions
  descriptions = soup_hotel.find('div', {'id':'property_description_content'}).text.strip().replace('\n',' ').split('.', 1)[1].strip()
  # Address
  address = soup_hotel.find('p', {'class':'address address_clean'}).text.strip().split('\n',1)[0]
  # User
  list_user = []
  for t in soup_hotel.find_all('div', {'class':'f9afbb0024 f0d4d6a2f5'}):
    username = t.text.strip()
    list_user.append(username)

  # Rating of hotel
  ratings = int(float(soup_hotel.find('div', {'class':'b5cd09854e d10a6220b4'}).text.strip().replace(',','.')))
  start = ratings-3
  stop = ratings+1 if ratings < 10 else 10

  list_url_hotels, list_locations, list_name_hotels, list_descriptions, list_address, list_usernames, list_ratings = [], [], [], [], [], [], []
  for username in list_user:
    list_url_hotels.append(url_hotel)
    list_locations.append(location)
    list_name_hotels.append(name_hotel)
    list_descriptions.append(descriptions)
    list_address.append(address)
    list_usernames.append(username)
    list_ratings.append(random.randint(start, stop))

  return list_url_hotels, list_locations, list_name_hotels, list_descriptions, list_address, list_usernames, list_ratings
````

````python
URL_Hotel, Locations, Name_Hotel, Descriptions, Address, User, Rating = [], [], [], [], [], [], []

for url in tqdm(list_hotels):
  try:
    data = Crawl_Hotels_Data(url)
    URL_Hotel.extend(data[0])
    Locations.extend(data[1])
    Name_Hotel.extend(data[2])
    Descriptions.extend(data[3])
    Address.extend(data[4])
    User.extend(data[5])
    Rating.extend(data[6])
  except:
    # print(f'Error -> {url}')
    continue
````
